```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Генерируем больше данных и называем узлы станциями метро
# Данные не имеют связи с действительностью
metro_data = {
    'station_from': ['Краснопресненская', 'Киевская', 'Курская', 'Лубянка', 'Таганская', 'Арбатская', 'Пушкинская', 'Смоленская', 'Чистые пруды', 'Тверская', 'Тверская'],
    'station_to': ['Тверская', 'Таганская','Киевская', 'Курская', 'Арбатская','Чистые пруды','Лубянка',    'Пушкинская', 'Пушкинская',  'Краснопресненская',  'Лубянка'],
    'distance': [2.5, 3.0, 1.5, 2.0, 3.5, 2.2, 2.8, 1.8, 3.3, 2.7, 4]
}


# Создаем DataFrame
metro_df = pd.DataFrame(metro_data)

# Создаем новый граф
metro_G = nx.Graph()

# Добавляем ребра в граф
for _, row in metro_df.iterrows():
    metro_G.add_edge(row['station_from'], row['station_to'], distance=row['distance'])

# Визуализируем граф метро
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(metro_G)
distances = nx.get_edge_attributes(metro_G, 'distance').values()
nx.draw(metro_G, pos, with_labels=True, node_color='lightgreen', node_size=1500, edge_color='gray', width=list(distances), edge_cmap=plt.cm.Greens)
plt.title("Сеть метро")
plt.show()

# Выводим таблицу
metro_df
```
